```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder,OneHotEncoder

df=pd.read_csv("./datasets./bank-additional-full.csv",sep=';')
df.head()
```

```python
df_pre=df['previous']
df_pre


###이미 수치형 데이터로 존재
```

```python
pout=df['poutcome']
pout
```

```python
pout= pout.apply(lambda x: x.replace('failure','0'))
pout = pout.apply(lambda x: x.replace('success','1'))
pout = pout.apply(lambda x: x.replace('nonexistent','2'))
```

```python
encorder=LabelEncoder()
encorder.fit(pout)
labels_p=encorder.transform(pout)
labels_p
# pd.DataFrame(labels,index=index,columns="poutcome")
```

```python
labels_p=labels_p.reshape(-1,1)
oh_encorder=OneHotEncoder()
oh_encorder.fit(labels_p)
oh_labels_p=oh_encorder.transform(labels_p)
oh_labels_p.toarray()
```
